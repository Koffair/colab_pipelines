## Vanilla inference

```python
model_path = "/content/gdrive/My Drive/Colab Notebooks/models/"
from huggingsound import SpeechRecognitionModel

model = SpeechRecognitionModel(model_path)
```

```python
import os
test_path = "/content/gdrive/My Drive/Colab Notebooks/test_data"
audio_paths = [os.path.join(test_path, f) for f in os.listdir(test_path) if os.path.isfile(os.path.join(test_path, f))]
```

```python
audio_paths
```

```python
from IPython.display import Audio
from IPython.display import display
for audio_file in audio_paths:
      wn = Audio(audio_file, autoplay=False)
      display(wn)
```

```python
transcriptions = model.transcribe(audio_paths)
```

```python
for transcript in transcriptions:
  print(transcript["transcription"])
```
